2a + b + c = 4  
a + 3b + 2c = 5  
a = 6

```python
import numpy as np
from numpy.linalg import solve, det, inv
```

```python
A = np.array([[2,1,1],[1,3,2],[1,0,0]])
b = np.array([4,5,6])
```

```python
x = solve(A,b)
if det(A) != 0:
    B = inv(A)
    y = B.dot(b)
```

```python
print(x)
print(y)
```

```python
points = np.arange(-5,5,0.01)
```

```python
print(points[:20])
```

```python
x,y = np.meshgrid(points,points)
```

```python
print(x[:10])
print(y[:10])
```

```python
z = np.sqrt(x**2 + y**2)
print(z)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,8))
plt.imshow(z, cmap=plt.cm.Blues)
plt.colorbar()
plt.title("Image plot of $\sqrt{x^2 + y^2}$ for a grid of values")
```
