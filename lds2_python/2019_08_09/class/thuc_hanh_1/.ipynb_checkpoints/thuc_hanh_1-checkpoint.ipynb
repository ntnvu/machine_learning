```python
import pandas as pd
```

```python
# doc cac sinh vien cua khoa CN, TO, VL
# doc ket qua hoc tap cua cac sinh vien nay
# tu ket qua hoc tap tinh ra diem trung binh cua cac sinh vien
# rap diem trung binh vao du lieu sinh vien
```

```python
df_cn = pd.read_csv('sinh_vien_CN.csv', dtype={'didong': 'S10'})
```

```python
print(df_cn.shape)
```

```python
print(df_cn.head())
```

```python
df_to = pd.read_csv('sinh_vien_TO.csv', dtype={'didong': 'S10'})
df_vl = pd.read_csv('sinh_vien_VL.csv', dtype={'didong': 'S10'})
```

```python
df = pd.concat([df_cn, df_to, df_vl])
df.index = range(df.shape[0])
```

```python
df_kqht = pd.read_csv('ket_qua.csv')
df_kqht
```

```python
# tinh diem trung binh
```

```python
df_dtb = df_kqht.groupby('masv', as_index=False).diem.mean()
```

```python
type(df_dtb)
```

```python
df_ = pd.merge(df, df_dtb, on='masv')
```

```python
df_
```

```python
# Cho biet cac sv co dtb >= 5
```

```python
df_[df_['diem'] >= 5]
```

```python
# Cho viet cac sinh vien khoa CN co dtb >= 5
```

```python
df_sort = df_[(df_['diem'] >= 8) & (df_['makh'] == 'CN')]
```

```python
df_sort
```

```python
df_sort.sort_values(by=['diem'],inplace=True, ascending=False)
```

```python
# cho biet cac sinh vien co diem tb nho nhat
```

```python
df_[df_['diem'] == df_['diem'].min()]
```

```python
# cho biet max, min, dtb trong moi khoa
```

```python
df_.groupby('makh')['diem'].max()
```

```python
df_.groupby('makh')['diem'].min()
```

```python
df_max_min = df_.groupby('makh')['diem'].agg(['min', 'max']).reset_index()
df_max_min
```

```python
df_max_min = df_.groupby('makh')['diem'].agg(['min', 'max'])
df_max_min['makh'] = df_max_min.index
df_max_min.index = range(df_max_min.shape[0])
df_max_min
```
