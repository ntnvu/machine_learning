```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = open('years.txt')
data = f.read()
f.close()
```

```python
years = [int(y) for y in data.split(',')]
years = pd.Series(years)
print(years)
```

```python
f = open('pops.txt')
data = f.read()
f.close()
```

```python
pops = [float(p) for p in data.split(',')]
pops = pd.Series(pops)
print(pops)
```

```python
plt.plot(years, pops)
plt.show()
```

```python
# ve cac nam tu 1950 den 1954
cond = (years>=1950) & (years<=1954)
x = years[cond]
y = pops[cond]
```

```python
print(x)
print(y)
```

```python
plt.plot(x,y,color='m',linestyle='dashed',marker='o',markersize=12, 
         linewidth=2)
plt.show()
```

```python
df = pd.read_excel("group_by_dep.xls")
```

```python
df
```

```python
x = df.index
y = df['sum_sal']
plt.figure(figsize=(14,5))
plt.plot(x,y)
plt.title('Thong ke luong theo phong ban')
plt.xlabel('Phong', color='red')
plt.ylabel('Tong luong', color='red')
plt.xticks(x, df['DEPARTMENT_NAME'])
plt.grid(True)
plt.show()
```

```python
x = df.index
y = df['count_emp']
plt.figure(figsize=(14,5))
plt.plot(x,y)
plt.title('Thong ke so nhan vien theo phong ban')
plt.xlabel('Phong', color='red')
plt.ylabel('Nhan vien', color='red')
plt.xticks(x, df['DEPARTMENT_NAME'])
plt.grid(True)
plt.show()
```

```python
y = df['count_emp']*1000
x = df.index
plt.figure(figsize=(14,5))

plot_1 = plt.plot(x,y, color='blue')

y_ = df['sum_sal']
plot_2 = plt.plot(x,y_,color='magenta')

plt.title('Thống kê tong luong va dem so nhan vien theo phòng', color='red')
plt.xlabel('Phòng', color='red')
plt.ylabel('Tong luong - So nhan vien', color='red')
plt.xticks(x, df['DEPARTMENT_NAME'])
plt.legend(['Số nhân viên', 'Tổng lương'],loc=1,title='Ghi chu')
plt.grid(True)
plt.show()
```
